#Part 2: Univariate Analysis

##Descriptive Statistics and Visualizations-

```python
# Relevant Descriptive Staitistics for Continous Variables:
df[['Age', 'Fare']].describe()
```

```python
# Histogram for continous variable

import matplotlib.pyplot as plt

df[['Age']].hist(bins=10, figsize=(10, 5))
plt.xlabel("Values")
plt.ylabel("Frequency")
df[['Fare']].hist(bins=10, figsize=(10, 5))

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Relevant Descriptive Staitistics for Categorical Variables:

print("Frequency counts of unique values in \n\n ", var, df[['Survived', 'Pclass', 'Sex', 'Embarked']].nunique())
```

```python
# Bar plot for categorical variables

import matplotlib.pyplot as plt
for column in ['Survived', 'Pclass', 'Sex', 'Embarked']:
  df[column].value_counts().plot(kind='bar', figsize=(10, 5))
  plt.title(f'Bar plot for {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

#Part 3: Bivariate Analysis

## Age vs Survived

```python


# Group data by 'Survived' and calculate the mean age for each group
age_survival = df.groupby('Survived')['Age'].mean()

# Create the bar plot
age_survival.plot(kind='bar', figsize=(10, 5))

plt.title('Average Age vs Survived')
plt.xlabel('Survived')
plt.ylabel('Average Age')
plt.xticks(rotation=0)
plt.show()
```

##Pclass vs Fare

```python
fare_class = df.groupby('Pclass')['Fare'].mean()

fare_class.plot(kind = 'bar', figsize=(10,5))

plt.title('Average Fare vs Pclass')
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.xticks(rotation=0)
plt.show()
```

## Sex vs Survived

```python


import matplotlib.pyplot as plt
survival_sex = df.groupby('Sex')['Survived'].mean()

survival_sex.plot(kind='bar', figsize=(10, 5))

plt.title('Survival Rate vs Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

##Survival Rate vs Pclass

```python

survival_pclass = df.groupby('Pclass')['Survived'].mean()

survival_pclass.plot(kind='bar', figsize=(10, 5))


plt.title('Survival Rate vs Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```
